$take$

- first argument: indices
- indices will take some values in array: in a list
- it tries to extract the data
- that values we provided represents columns or rows??
- axis=0 represent about rows
- axis=1 represent about columns
- by default axis mentioned as zero

```python
# case-1
visa_df.take([100,200,300])
# it retrives 3 rows which having index 100,200,300
```

```python
# case-2
visa_df.take([100,200,300],axis=1)
# now this will check for 100th column, 200th column and 300th column
# But originally we have 12 column (0,11)
# will get error
```

```python
# case-3: select 3 columns 2nd, 5th and 7th columns
# using take
visa_df.take([2,5,7],axis=1)
# the output also behave as dataframe
```
